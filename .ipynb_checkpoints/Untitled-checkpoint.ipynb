```python
import pandas
import numpy
from matplotlib import pyplot as plt
import seaborn
plt.style.use('ggplot')
seaborn.set_style('dark',rc={'axes.labelcolor':'white','xtick.color':'white','ytick.color':'white'})
%matplotlib notebook
import datetime
pandas.set_option('display.float_format', '{:20,.2f}'.format)   #suppressing scientific notation
idx = pandas.IndexSlice
```

```python
test = pandas.read_excel('paradox_charts.xlsx','Sheet1')
```

```python
fig, ax = plt.subplots()
ax.grid()
ax2 = ax.twinx()
ax.plot('TOPIX Index',data=test,c='r')
ax.plot('TOPIX Index future',data=test, c='y')
ax2.plot('FTSE 100',data=test,c='b')
ax.legend(['TOIX Index','TOPIX Index future'])
ax2.legend(['FTSE 100'])
ax.set_xlabel('Timestamps')
```
